The purpose of this analysis was to determine if there was a relationship between amount of steps taken in a day, and calories burnt, based off data from fitbit. The dataframe consisted of 457 dat points, tracking how many calories were burnt in a single day, and how many steps were taken in that day.

The data was visualized (as seen below) and shows there is a clear positive correlation between steps taken in a day, and calories burnt.


```python
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

# Set your folder path (adjust capitalization/spaces exactly as in Drive)
folder_path = '/content/drive/My Drive/Fit Bit Project/'

# Load CSV files
import pandas as pd

activity = pd.read_csv(folder_path + 'dailyActivity_merged.csv')

activity.info()
```

```python
# Simple clean style
sns.set(style='whitegrid', context='talk')

plt.figure(figsize=(10,6))
sns.regplot(
    x='TotalSteps',
    y='Calories',
    data=activity,
    scatter_kws={'alpha':0.5, 's':50, 'color':'steelblue'},
    line_kws={'color':'darkred', 'lw':2}
)

plt.title('Steps vs Calories Burned', fontsize=18, fontweight='bold')
plt.xlabel('Total Steps', fontsize=14)
plt.ylabel('Calories Burned', fontsize=14)
plt.tight_layout()
plt.show()
```

Found a strong positive correlation between total steps and calories burned
